```python
from helldivers2.load_data.load_data import (
    load_data
)

from helldivers2.preprocessing.preprocessing import (
    aggregate_total_players
)

df_raw = load_data()

ts_players = aggregate_total_players(
    df_raw
)

from statsmodels.graphics.tsaplots import (
    plot_acf,
    plot_pacf
)

plot_acf(
    ts_players["total_players"],
    lags=30
)

plot_pacf(
    ts_players["total_players"],
    lags=30
)
```

```python
EXPERIMENT_NAME = "only lags"

MODEL_TYPES = [
    "linear_regression",
    #"random_forest",
    #"xgboost"
]

all_results = []

for model_type in MODEL_TYPES:

    

    FEATURE_GROUPS = {
        "lags": True,
        "rolling": True,
        "time": True,
        "momentum": True,
        "ema": True,
        "extremes": True,
        "interactions": True,
        "warbonds": True,
        "fourier": True
        
    }

    # FEATURES = [
    #     "lag_1",
    #     "lag_24",
    #     "lag_168",
    #     #"rolling_mean_24"
    # ]

    results = run_experiment(

        experiment_name=EXPERIMENT_NAME,

        model_type=model_type,

        features=FEATURES,

        feature_groups=FEATURE_GROUPS,

        use_log_target=False
    )

    all_results.append({

        "model": results["model_name"],

        "rmse": results["metrics"]["cv_rmse_mean"],

        "mae": results["metrics"]["cv_mae_mean"]
    })

results_df = pd.DataFrame(
    all_results
)

results_df
```

```python
import matplotlib.pyplot as plt
results_df = pd.DataFrame({

    "timestamp": results["timestamps"],

    "actual": results["y_test"],

    "predicted": results["predictions"]
})

# =========================================
# FILTER TO MAY 13 -> MAY 15
# =========================================

results_subset = results_df[
    (results_df["timestamp"] >= "2026-05-13") &
    (results_df["timestamp"] < "2026-05-16")
].copy()

# =========================================
# PLOT
# =========================================

plt.figure(figsize=(16, 6))

plt.plot(
    results_subset["timestamp"],
    results_subset["actual"],
    label="Actual"
)

plt.plot(
    results_subset["timestamp"],
    results_subset["predicted"],
    label="Predicted"
)

plt.title("Hourly Forecast: May 13 - May 15")

plt.xlabel("Timestamp")
plt.ylabel("Players")

plt.xticks(rotation=45)

plt.legend()

plt.show()
```
